## Calculate KPI's
What is churn rate?

```python
churn_rate = (df["Exited"].mean()) * 100
print(churn_rate)
```

## Churn Count Visualization
For this visualization countplot is considered beacuse it Counts categories

```python
plt.figure(figsize = (5,4))

sns.countplot(x = "Churn_Status", hue = "Gender", data = df)
plt.title("Customer Churn Count")
plt.show()
```

## Age Distribution 
For this visualization histplot is considered beacuse it indicates Distribution of numbers

Are customers mostly young or old?

```python
plt.figure(figsize = (7,5))

sns.histplot(df["Age"], bins = 20)
plt.title("Age Distribution")

plt.show()
```

## Age vs Churn
For this visualization Boxplot is considered beacuse boxplot Detect spread and outliers

Do older customers churn more?

```python
plt.figure(figsize = (7,5))

sns.boxplot(x = "Churn_Status", y = "Age", data = df)
plt.show()
```

## Geography Analysis
Which geography churns most?

```python
geo_churn = pd.crosstab(df["Geography"],df["Churn_Status"])

geo_churn.plot(kind = "bar")
plt.title("Geography Analysis")
plt.show()
```

## Active Customers Analysis
Do inactive customers churn more?

```python
pd.crosstab(df["IsActiveMember"],df["Churn_Status"],normalize="index") * 100
```

## Products Analysis
Does number of products affect churn?

```python
sns.countplot(
    x="NumOfProducts",
    hue="Churn_Status",
    data=df
)

plt.show()
```
