```python
# Removing Unnecessary Columns but in this case we remove only one columns but in case you have to
# remove many columns this is the method in drop_cols you can add multiple columns names....

drop_cols = [
    "RowNumber"
]

df = df.drop(columns=drop_cols)
df.head(5)
```

```python
df.info()
```

```python
df.shape
```

```python
# create a new column churn_status

df["Churn_Status"] = df["Exited"].map({ 0 : "Retained", 1:"Churned"})
```

```python
df.head()
```

```python
# create a new column age_group

df["Age_Group"] = pd.cut(df["Age"], 
                         bins = [18,30,40,50,60,100],
                         labels = [
                             "18-30",
                             "31-40",
                             "41-50",
                             "51-60",
                             "60+"
                         ], include_lowest=True
                        )
```

```python
df.head()
```

```python
df[df['Age'] == 18]
```

```python
# Renaming columns according to snake casing for better readability and documentation
df.columns = df.columns.str.lower()
df = df.rename(columns={'customerid':'customer_id',
                        'creditscore':'credit_score',
                        'numofproducts':'num_of_products',
                        'hascrcard':'has_cr_card',
                        'isactivemember':'is_active_member',
                        'estimatedsalary':'estimated_salary',
                       })
```

```python
# Save Cleaned Data to csv

df.to_csv("D:/ABHAY/Projects/Bank_Customer_Churn_Project/data/cleaned/churn_cleaned.csv")
```
